```python
#Links the DataSet
dataframe = pd.read_csv("KAG_conversion_data.csv")
```

```python
#Taking a look at the data
print(dataframe)
```

```python
# How large is each gender group?
gender_femles= (dataframe['gender'] == 'F')
gender_males= (dataframe['gender'] == 'M')
print('Female Sample Size:', len(dataframe[gender_femles]))
print('Male Sample Size:', len(dataframe[gender_males]))
print('Entire population sample size:' , len(dataframe))
```

```python
#Statitical description of male data 
print(dataframe[gender_males].describe())
```

```python
#Statitical description of femles data 
print(dataframe[gender_femles].describe())
```

```python
#Function will create a box plot per column for each gender
def box_plot(data1,data2,data3,data4):
    plot_gender = [data1, data2]
    plt.boxplot(plot_gender)
    plt.xlabel(data4)
    plt.ylabel(data3)
    plt.show()
    
```

```python
#Comparison of genders on ad clicks
box_plot(dataframe[gender_males]['Clicks'],dataframe[gender_femles]['Clicks'],"Number of Clicks","Gender")
```

```python
#Comparison of genders on ad interest
box_plot(dataframe[gender_males]['interest'],dataframe[gender_femles]['interest'],"Interest","Gender")
```

```python
#Comparison of the money spent per gender to generate ads. 
box_plot(dataframe[gender_males]['Spent'],dataframe[gender_femles]['Spent'],"Spent","Gender")
```

```python
#Comparison of total approved conversions
box_plot(dataframe[gender_males]['Approved_Conversion'],dataframe[gender_femles]['Approved_Conversion'],"Approved_Conversion","Gender")
```
